# Reflect Tables into SQLAlchemy ORM

```python
# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, inspect, func
```

```python
# create engine to hawaii.sqlite
engine = create_engine("sqlite:///Resources/hawaii.sqlite")
```

```python
# reflect an existing database into a new model
base = automap_base()

# reflect the tables
base.prepare(engine, reflect=True)
```

```python
# View all of the classes that automap found
base.classes.keys()
```

```python
# Save references to each table
measurement = base.classes.measurement 
station = base.classes.station
#print(Base.classes.keys())
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
```

```python
# TEST Get the table names using `inspect()`.
inspector = inspect(engine)
inspector.get_table_names()
```

```python
# TEST Get a list of column names and types
columns = inspector.get_columns('station')
for c in columns:
    print(c['name'], c["type"])
```

```python
#test1 measurements
for row in session.query(measurement.station,measurement.date, measurement.prcp ).limit(5).all():
    print(row)
```

```python
#test2 measurements
measurements = session.query(measurement).order_by(measurement.station).count()
print(measurements)
```

```python
#test1 station
for row in session.query(station.station,station.name).all():
    print(row)
```

# Exploratory Precipitation Analysis

```python
# Find the most recent date in the data set.
most_recent_date = session.query(measurement.date).order_by(measurement.date.desc()).first()
print (most_recent_date)
```

```python
# Design a query to retrieve the last 12 months of precipitation data and plot the results. 
# Starting from the most recent data point in the database. 
date_string = most_recent_date[0]
#print (date_string) -> 2017-08-23

# Calculate the date one year from the last date in data set.
date_obj = dt.datetime.strptime(date_string, '%Y-%m-%d')
year_days_before = date_obj - dt.timedelta(days = 365)
formatted_year_ago = year_days_before.strftime('%Y-%m-%d')
#print(year_days_before) -> 2016-08-23 00:00:00
#print(formatted_year_ago) ->-> 2016-08-23

# Perform a query to retrieve the data and precipitation scores
data_prec_query = session.query(measurement.date, measurement.prcp).\
            filter(measurement.date >= formatted_year_ago).all()


# Save the query results as a Pandas DataFrame. 
data_prec_df = pd.DataFrame(data_prec_query)

#set index
data_prec_df = data_prec_df.set_index('date')

# Sort the dataframe by date
data_prec_df = data_prec_df.sort_values(by = ['date'])

# Use Pandas Plotting with Matplotlib to plot the data
data_prec_df.plot(figsize = (7,5), rot=90)
plt.xlabel('Date') #Set X-axis Label
plt.ylabel('Inches') #Set Y-axis Label
plt.legend(['precipitation'], loc="upper center",fontsize=15)
```

```python
# Use Pandas to calculate the summary statistics for the precipitation data
data_prec_df.describe()
```

# Exploratory Station Analysis

```python
# Design a query to calculate the total number of stations in the dataset
station_count_all = session.query(station).order_by(station.station).count()
print(station_count_all)
#[(9,)]
```

```python
# Design a query to find the most active stations (i.e. which stations have the most rows?)
# List the stations and their counts in descending order.

# This JOINs the data in the two tables together into a single dataset (here in the form of a tuple).
sel = [station.station, func.count(measurement.station)]
session.query(*sel).filter(station.station == measurement.station).\
    group_by(measurement.station).\
    order_by(func.count(measurement.station).desc()).all()
```

```python
# Using the most active station id from the previous query, calculate the lowest, highest, and average temperature.

station_count = session.query(*sel).filter(station.station == measurement.station).\
    group_by(measurement.station).\
    order_by(func.count(measurement.station).desc()).all()
# Get the first value of the tuple station_count[0]
stations = [t[0] for t in station_count] # I will need stations[0] = 'USC00519281' below

#Calculate:
sel1 = [func.min(measurement.tobs), func.max(measurement.tobs), func.avg(measurement.tobs)]
session.query(*sel1).filter(measurement.station == stations[0]).all()
```

```python
# Using the most active station id
# Query the last 12 months of temperature observation data for this station and plot the results as a histogram

year_temperature_data = session.query(measurement.tobs).filter_by(station = stations[0]).\
    filter(measurement.date >= formatted_year_ago).all()   

#let's create a dataframe to plot data
year_temperature_data_df = pd.DataFrame(year_temperature_data)

#Plot the last 12 months of temperature observation data
year_temperature_data_df.plot(kind = 'hist', bins = 12)
plt.xlabel('Temperature')
plt.ylabel('Frequency')
#Text(0.5, 4.183333333333314, 'Temperature')
```
